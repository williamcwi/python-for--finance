# Compare stock performance with Python

## 01 Define stocks to compare from Yahoo Finance

```python
stocks = 'XRP-USD, ADA-USD, DOGE-USD, MATIC-USD'
```

```python
print(stocks)
```

## 02 Scrape stock data from yfinance API

```python
# pip install yfinance

import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as pyplot

import yfinance as yf
```

```python
data = yf.download(stocks)
```

```python
data
# data.info()
```

## 03 Visualise stock performance

```python
data['Close']
```

```python
pyplot.rcParams['figure.figsize'] = (20, 10)
pyplot.rcParams.update({'font.size':20})

data['Close'].plot()
```
